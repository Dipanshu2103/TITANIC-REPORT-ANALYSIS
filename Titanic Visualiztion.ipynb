## Create a pie chart which should show the distribution of male and female

```python
a= df['Sex'].value_counts()
```

```python
plt.pie(a, labels=a.index, autopct="%1.1f%%",colors=['skyblue','pink'], startangle=90, explode=[0,0.1])
plt.title("Sex Proportion")
```

```python
def number(pct): 
    return f"{int(round(pct*sum(a)/100))}"

plt.pie(a, labels=a.index, colors=['skyblue','pink'], startangle=90, autopct=number)
plt.title("Sex Proportion")
plt.show()
```

## What is the distribution of ages?

```python
count, bins,_ =plt.hist(df['Age'],bins=8,rwidth=0.8)
labels = [f'{int(bins[i])}-{int(bins[i+1])}' for i in range(len(bins)-1)]
plt.xticks([(bins[i] + bins[i+1])/2 for i in range(len(bins)-1)],labels)
plt.title('Age Distribution')
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
plt.show()
```

### with seaborn

## Did gender affect the survival rate?

```python
a = df[df['Sex']=='male']['Sex'].value_counts()
b = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]['Sex'].value_counts()
c = round((b/a)*100,2)
c
```

```python
d = df[df['Sex']=='female']['Sex'].value_counts()
e = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]['Sex'].value_counts()
f = round((e/d)*100,2)
f
```

```python
survival_rates = {
    'Male': float(c['male']),
    'Female': float(f['female'])}
survival_rates
```

```python
plt.bar(survival_rates.keys(), survival_rates.values(), color=['skyblue', 'pink'])
plt.ylabel("Survival Percentage")
plt.title("Survival Rate by Gender")
plt.show()
```

```python
plt.pie(survival_rates.values(),
        labels=survival_rates.keys(),
        autopct='%1.1f%%',
        colors=['skyblue', 'pink'],
        startangle=90)

plt.title("Survival Rate by Gender")
plt.axis('equal')
plt.show()
```
